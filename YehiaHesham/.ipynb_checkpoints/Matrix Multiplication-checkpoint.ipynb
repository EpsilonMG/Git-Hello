```python
#Methods Declaration

def input_matrix():
    matrix = []
    try:
        n = int(input("Enter N for N x N matrix: "))
    except:
        print("wrong input, please try again")
        return False
    for i in range(n):
        row = []
        for j in range(n):
            try:
                row.append(int(input(f'Item at ({i},{j}) :')))
            except:
                print("wrong input, please try again")
                return False
        matrix.append(row)
    return matrix


def display_matrix(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(n):
            print(matrix[i][j], end = " ")
        print()

        
def multiply_matrix(matrix_1,matrix_2):
    
    if not len(matrix_1) == len(matrix_2):
        print("2 matrix must have the same size")
        return False

    result = []
    n = len(matrix_1)
    for i in range(n):
        row = []
        for j in range(n):
            row.append(0)
        result.append(row)
    
    for i in range(len(matrix_1)):
        for j in range(len(matrix_2[0])):
            for k in range(len(matrix_2)):
                result[i][j] += matrix_1[i][k] * matrix_2[k][j]
                
    return result
        
    
```

```python
#Program

matrix_1 = input_matrix()
print("First Matrix: ")
display_matrix(matrix_1)

matrix_2 = input_matrix()
print("Second Matrix: ")
display_matrix(matrix_2)

matrix_multiplication = multiply_matrix(matrix_1,matrix_2)
print("The value of multiplication of the 2 matrix is: ")
display_matrix(matrix_multiplication)
```
