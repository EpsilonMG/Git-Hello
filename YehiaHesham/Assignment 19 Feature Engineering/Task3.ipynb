### 1) Using titanic_full.csv dataset, extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', then plot a countplot
showing the count for each title in the dataset.

```python
df = pd.read_csv("datasets/titanic_full.csv")
```

```python
df['Title'] = [name.split(",")[-1].split(".")[0] for name in df.Name]
```

```python
df[['Name','Title']].sample(10)
```

```python
df['Title'].value_counts()
```

### Saving Dataset

```python
df.to_csv("datasets/edited_titanic_full.csv",index=False)
```

### 2) Using sendy_logistics.csv extract user_id in a seperate column.

```python
df = pd.read_csv("datasets/sendy_logistics.csv")
```

```python
df['User Id'][0:5]
```

<font color = 'blue'>The column already seems to exist, but we can clean it up a bit </font>

```python
df['User Id'] = [user_id.split("_")[-1] for user_id in df['User Id'] ]
```

```python
df['User Id']
```

<font color = 'blue'> lets clean up other columns that have the same problem</font>

```python
df.head()
```

```python
df['Order No'] = [order_no.split("_")[-1] for order_no in df['Order No'] ]
df['Order No'][0:5]
```

```python
df['Rider Id'] = [rider_id.split("_")[-1] for rider_id in df['Rider Id']]
df['Rider Id']
```

```python
df.head()
```

<font color = 'blue'> much better.</font>

### Saving Dataset

```python
df.to_csv("datasets/edited_sendy_logistics2.csv",index=False)
```
