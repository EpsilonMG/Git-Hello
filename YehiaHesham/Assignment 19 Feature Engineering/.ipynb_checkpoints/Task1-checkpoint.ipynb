```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

sns.set(rc = {'figure.figsize':[9,9]},font_scale= 1.2)
```

### 1) Using sendy_logistics.csv dataset create a new column and call it 'Temp_Diff' that calculates the difference of 'Tempreture' Column and current temperture.
<font color = 'blue'>note: you can get current tempreture from searching on google.</font>

```python
df = pd.read_csv("datasets/sendy_logistics.csv")
```

```python
df.keys()
```

```python
df['Temperature']
```

```python
df['Temp_Diff'] = df['Temperature'] - 18
```

```python
sns.kdeplot(x='Temperature',data = df, color = 'blue')
sns.kdeplot(x='Temp_Diff',data = df, color = 'red')
plt.show()
```

### 2) Using sendy_logistics.csv dataset create a new column and call it 'Fehr_Temp' that calculates the fahrenheit tempreture from celsius 'Tempreture' Column.
<font color = 'blue'> note: you can get the converting formula celsius to fahrenheit from google. </font>


```python
df['Fehr_Temp'] = ( df['Temperature'] * (9.0/5) ) + 32
```

```python
sns.kdeplot(x = 'Temperature', data = df, color = 'red')
sns.kdeplot(x = 'Fehr_Temp', data = df, color = 'blue')
plt.xlabel("Temperature")

red_patch = mpatches.Patch(color='red', label='celsius')
blue_patch = mpatches.Patch(color='blue', label='fahrenheit')
plt.legend(handles=[red_patch, blue_patch])

plt.show();
```
