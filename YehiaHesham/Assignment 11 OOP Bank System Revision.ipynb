```python
import re
#=======================================================================
#==== BLOCK 1: DECLARATIONS OF USER CLASS ==============================
#=======================================================================


#=================================================== PARENT CLASS (USER)
class User ():

    #=================================================== CONSTRUCTOR
    def __init__ (self, name = "No Name", age = 0, gender = "No Gender"):
        self.name = name
        self.age = age
        self.gender = gender
      
    #=================================================== SETTERS AND GETTERS
    def set_name(self, name):
        if type(name) is str:
            self.name = name
        else: 
            self.name = False
            
    def get_name(self):
        return self.name
    
    def set_age(self,age):
        if type(age) is int and age >= 18 and age <= 100:
            self.age = age
        else: 
            self.age = False
        
    def get_age(self):
        return self.age
    
    def set_gender(self,gender):
        if type(gender) is str:
            if gender.upper() in ['MALE','M']:
                self.gender = 'M'
            if gender.upper() in ['FEMALE','F']:
                self.gender = 'F'
        else: 
            self.gender = False
      
    def get_gender(self):
        return self.gender
    
    #=================================================== METHODS  
    def show_Details (self):
        return [self.name, self.age, self.gender]
```
