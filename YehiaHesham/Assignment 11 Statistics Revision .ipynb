```python
#another solution to calucate the mode of YearsExperience/Salary
mode = c.Counter(data.T)
test = mode.most_common()
data[data.index==test[0][0]]
```

```python
print("YearsExperience Correlation Coefficient: ", np.corrcoef(data['YearsExperience']))
print("Salary Correlation Coefficient: ", np.corrcoef(data['Salary']))
```

```python
#If you don't transpose the 2d array, the method interprets row as vector, so in your first few cases,
#it calculates the correlation coefficients of all the combinations of the rows, since each row is length 2 vector,
#you get all 1 as coefficients, which would happen when you try to fit a line with two points (always perfect fit).
#src:
#    https://stackoverflow.com/questions/44333504/np-corrcoef-behavior-with-pandas-dataframes

print("YearsExperience/Salary Correlation Coefficient of first 10 entries: \n") 
np.corrcoef(data[['YearsExperience','Salary']])
```
